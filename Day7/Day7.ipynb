```python
box_plot = df.Age
plt.boxplot(box_plot)
plt.title('Age')
plt.show()
```

```python
box_plot = df.MonthlyIncome
plt.boxplot(box_plot)
plt.title('Monthly Income')
plt.show()
```

- __Monthly Income is Rightskewed with several outliers.__

```python
box_plot = df.YearsAtCompany
plt.boxplot(box_plot)
plt.title('Years at Company')
plt.show()
```

```python
x = df.Age
y = df.MonthlyIncome
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.title('Age vs Monthly Income')
plt.show()
```
